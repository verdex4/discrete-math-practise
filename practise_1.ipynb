## Первая практика: проверка свойств бинарных отношений

Свойства:
- рефлексивность
- антирефлексивность
- симметричность
- антисимметричность
- транзитивность
- линейность

```python
def check_reflexivity(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if i == j and matrix[i][j] == 0:
                return False
    return True

def check_antireflexivity(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if i == j and matrix[i][j] == 1:
                return False
    return True

def check_symmetry(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == 1 and matrix[j][i] == 0:
                return False
    return True

def check_antisymmetry(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if i != j and matrix[i][j] == 1 and matrix[j][i] == 1:
                return False
    return True

def check_transitivity(matrix):
    # a = i
    # b = j
    # c = k
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == 1: # есть aRb
                for k in range(len(matrix[j])):
                    if matrix[j][k] == 1 and matrix[i][k] == 0: # есть bRc, но нет aRc
                        return False # противоречие, т.к. должно быть для всех a,b,c
    return True

def check_linearity(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if i != j and matrix[i][j] == 0 and matrix[j][i] == 0:
                return False
    return True
```

```python
import import_ipynb
from template import get_matrix

matrix = get_matrix()

print("Исходная матрица")
print(matrix)

print("Выполняются свойства")
print(f"Рефлексивность: {check_reflexivity(matrix)}")
print(f"Антирефлексивность: {check_antireflexivity(matrix)}")
print(f"Симметричность: {check_symmetry(matrix)}")
print(f"Антисимметричность: {check_antisymmetry(matrix)}")
print(f"Транзитивность: {check_transitivity(matrix)}")
print(f"Линейность: {check_linearity(matrix)}")
```
