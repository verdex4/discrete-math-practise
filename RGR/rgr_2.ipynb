## Расчетно-графическая работа 2. Топологическая сортировка

```python
U = {989, 1, 567, 2, 3, 9, 394, 240, 270, 999}
U
```

```python
def top_sort(U):
    while U:
        m = min(U)
        yield m
        U.remove(m)
```

```python
sorted_U = []
for x in top_sort(U):
    sorted_U.append(x)
print(sorted_U)
```
