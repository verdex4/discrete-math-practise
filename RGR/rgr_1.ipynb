## Расчетно-графическая работа 1. Выполнение операций над булевыми матрицами

```python
import numpy as np

A = np.random.randint(0,2,(3,3)) # случайная булева матрица 3x3
B = np.random.randint(0,2,(3,3))

print("A")
print(A)

print("B")
print(B)
```

```python
print("Дизъюнкция: A U B")
A_or_B = np.logical_or(A, B).astype(int)
print(A_or_B)
```

```python
print("Транпонирование: A^T")
A_T = np.transpose(A)
print(A_T)
```

```python
def invert(matrix):
    return np.subtract(np.ones(shape=matrix.shape), matrix).astype(int)

print("Инвертирование: 1 - A")
bar_A = invert(A)
print(bar_A)
```

```python
print("Вычитание: A - B")
A_sub_B = A & invert(B)
print(A_sub_B)
```

```python
print("Умножение: A x B")
AxB = (A @ B > 0).astype(int)
print(AxB)
```
